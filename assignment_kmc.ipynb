**Q1.** This question is a case study for $k$ means clustering.

1. Load the `airbnb_hw.csv` data. Clean `Price` along with `Beds`, `Number of Reviews`, and `Review Scores Rating`.
2. Maxmin normalize the data and remove any `nan`'s (`KMeans` from `sklearn` doesn't accept `nan` input).
3. Use `sklearn`'s `KMeans` module to cluster the data by `Beds`, `Number of Reviews`, and `Review Scores Rating` for `k=6`.
4. Use `seaborn`'s `.pairplot()` to make a grid of scatterplots that show how the clustering is carried out in multiple dimensions.
5. Use `.groupby` and `.describe` to compute the average price for each cluster. Which clusters have the highest rental prices?
6. Use a scree plot to pick the number of clusters and repeat steps 4 and 5.

````python
# PART 1
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/DS3001/kmc/main/data/airbnb_hw.csv')

df['price'] = df['Price']
df['beds'] = df['Beds']
df['reviews'] = df['Number Of Reviews']
df['score'] = df['Review Scores Rating']

newdata = df.loc[:,['price','beds','reviews','score']]



newdata['price'] = newdata['price'].str.replace(',','')
newdata['price'] = pd.to_numeric(newdata['price'],errors='coerce')

newdata['beds'] = newdata['beds'].fillna(1)

newdata = newdata.dropna() # removing missing values

newdata.describe()
````

````python
# PART 2

def maxmin(x): # normalizing data
    norm = (x-min(x))/(max(x)-min(x))
    return norm

newdata2 = newdata.drop('price',axis=1)
newdata2 = newdata2.apply(maxmin)
````

````python
# PART 3
from sklearn.cluster import KMeans

model = KMeans(n_clusters=6, max_iter=100, n_init = 10, random_state=0) # clusering
model = model.fit(newdata2)
newdata2['cluster'] = model.labels_

newdata2.describe()
````

````python
# PART 4
import seaborn as sns

sns.pairplot(data=newdata2,hue='cluster') # scatterplot grid to show clustering in multiple dimentions
````

````python
# PART 5

newdata['cluster'] = model.labels_
newdata.loc[:,['price','cluster'] ].groupby('cluster').describe()


# explanation: overall, cluster 1 seems to have the highest rental prices with a mean and 50% much higher
# than the other clusters. One or a few big outliers appear to be skewing cluster 0 since it has a max value
# that is double the next closest max value for a cluster
````

````python
# PART 6
# scree plot repeating 4+5
import numpy as np
import seaborn as sns

k_bar = 15
k_grid = np.arange(1,k_bar+1)
SSE = np.zeros(k_bar)
for k in range(k_bar):
    model = KMeans(n_clusters=k+1, max_iter=100, n_init = 10, random_state=0) # model for k
    model = model.fit(newdata2)
    SSE[k] = model.inertia_ # Record the sum of squared error for k in the SSE vector

sns.lineplot(x=k_grid, y=SSE).set_title('Scree Plot') # Create scree plot
````

````python
# using 2 for number of clusters
model = KMeans(n_clusters=2, max_iter=100, n_init = 10, random_state=0) # k of 2
model = model.fit(newdata2)
newdata2['cluster'] = model.labels_

# part 4 repeat
sns.pairplot(data=newdata2,hue='cluster')

# part 5 repeat
newdata['cluster'] = model.labels_
newdata.loc[:,['price','cluster'] ].groupby('cluster').describe()

# cluster 0 is again skewed by a big outlier but still cluster 1
# has a significantly higher mean and 50th percentile value for price
````

2. Make scatterplots of the $(X1,X2)$ points by group for each of the datasets. As the `noise` goes up from 0.125 to 2.0, what happens to the visual distinctness of the clusters?
3. Create a scree plot for each of the datasets. Describe how the level of `noise` affects the scree plot (particularly the presence of a clear "elbow") and your ability to definitively select a $k$.
4. Explain the intuition of the elbow, using this numerical simulation as an example.

````python
# PART 1

import numpy as np
import pandas as pd

def createData(noise,N=50):
    np.random.seed(100) # Set the seed for replicability
    # Generate (x1,x2,g) triples:
    X1 = np.array([np.random.normal(1,noise,N),np.random.normal(1,noise,N)])
    X2 = np.array([np.random.normal(3,noise,N),np.random.normal(2,noise,N)])
    X3 = np.array([np.random.normal(5,noise,N),np.random.normal(3,noise,N)])
    # Concatenate into one data frame
    gdf1 = pd.DataFrame({'x1':X1[0,:],'x2':X1[1,:],'group':'a'})
    gdf2 = pd.DataFrame({'x1':X2[0,:],'x2':X2[1,:],'group':'b'})
    gdf3 = pd.DataFrame({'x1':X3[0,:],'x2':X3[1,:],'group':'c'})
    df = pd.concat([gdf1,gdf2,gdf3],axis=0)
    return df

df0_125 = createData(0.125)
df0_25 = createData(0.25)
df0_5 = createData(0.5)
df1_0 = createData(1.0)
df2_0 = createData(2.0)
````

````python
# PART 3
import matplotlib.pyplot as plt

# normalize
def maxmin(x):
    norm = (x-min(x))/(max(x)-min(x))
    return norm
# scree plot for each of the datasets to describe how noise affects the scree plot
def scree(data):
    X = data.loc[ : , ['x1','x2'] ]
    X = X.apply(maxmin)
    kbar = 15
    k_grid = np.arange(1,kbar+1)
    SSE = np.zeros(kbar)
    for k in range(kbar):
        model = KMeans(n_clusters=k+1, max_iter=100, n_init = 10, random_state=0)
        model = model.fit(X)
        SSE[k] = model.inertia_
    scree_plot, axes = plt.subplots()
    sns.lineplot(x=k_grid, y=SSE).set_title('Scree Plot')
    axes.set_ylim(0, 30)
````
